```python
import matplotlib.pyplot as plt
from collections import Counter
import re

words = re.findall(r'\b\w+\b', ftext.lower())
word_counts = Counter(words)

most_common_words = word_counts.most_common(10) 
labels, counts = zip(*most_common_words)

plt.figure(figsize=(10, 6))
plt.bar(labels, counts, color='blue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 Most Common Words in Text')


plt.show()
```

```python
# More operations with token
# Count the word "is"
token_to_find = 'is'
is_ = ftokens.count(token_to_find)
print(f"The token '{token_to_find}' appears {is_} times.")

# Count the word Hamlet
token_to_find = 'Hamlet'
hamlet = ftokens.count(token_to_find)
print(f"The token '{token_to_find}' appears {is_} times.")


#Plot a bar Chart
import matplotlib.pyplot as plt
labels = ['is', 'Hamlet']
counts = [is_, hamlet]

plt.bar(labels, counts, color=['blue', 'green'])
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Frequency of "is" and "Hamlet" in Text')

# Step 4: Display the chart
plt.show()
```

```python
# More operations with tokens

#Change token
new_text = ftext.replace('Hamlet', 'Vassily')
# Write the processed text back to a new file
with open('processed_data.txt', 'w') as file:
    file.write(new_text)
```
