```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Cust_Segmentation.csv")
df
```

```python
id = df["Customer Id"]
income = df["Income"]
```

```python
plt.plot(id, income)
plt.xlabel('Customer ID number')
plt.ylabel('Income')
plt.title('Customer ID num vs Income')
plt.grid(True)
plt.show()
```

```python
#Observing the line plot above, it seems like there is no relationship between the income and the customer id. 
```
