```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("FuelConsumption.csv")
df
```

```python
df.plot(x = "CO2EMISSIONS", y = "CYLINDERS", kind = "scatter", title = "fuel consumption in city vs # cylinders")
```

```python
# It seems like there is a correation between the fuel consumption in city and the number of cylinders a car has
df[["FUELCONSUMPTION_CITY", "CYLINDERS"]].corr()
```

```python
# It seems like there is a correation between the fuel consumption on the highway and the number of cylinders a car has
# However it is less obvious than the previous correlation
df[["FUELCONSUMPTION_HWY", "CYLINDERS"]].corr()
```

```python
sns.boxplot(x = "TRANSMISSION", y = "FUELCONSUMPTION_CITY", data=df)
plt.title("Relation between transmission and fuel consumption")
```
