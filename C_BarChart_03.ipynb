```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("drinks.csv")
df
```

```python
avg_beer_servings = df.groupby('continent')['beer_servings'].mean()
```

```python
avg_beer_servings.plot(kind='bar', title='Average beer serving per Continent')
plt.xlabel('Continent')
plt.ylabel('Count')
plt.grid(True)
```

```python
#We can observe that Europe is the continent with the most beer serving
# I used matplotlib and pandas plot for the bar chart above
```
