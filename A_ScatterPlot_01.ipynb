```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Import the data
df = pd.read_csv("drinks.csv")
```

```python
#Select the variable we are interested in
total_alcohol = df["total_litres_of_pure_alcohol"]
wine = df["wine_servings"]
```

```python
plt.scatter(wine, total_alcohol)
plt.xlabel('Wine consumption')
plt.ylabel('Total alcohol comsuption')
plt.title('Wine vs Total alcohol comsuption ')
plt.grid(True)
plt.show()
```
